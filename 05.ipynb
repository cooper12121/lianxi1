```python
"""
为标记分类创建自定义模型：主体+头部 
 """
import torch.nn as nn
from transformers import XLMRobertaConfig
from transformers.modeling_outputs import TokenClassifierOutput
from transformers.models.roberta.modeling_roberta import RobertaModel,RobertaPreTrainedModel
class XLMrobertaForTokenClassficiation(RobertaPreTrainedModel):
    config_class = XLMRobertaConfig
    def __init__(self, config):
        super().__init__(config)
        self.num_labels = config.num_labels
        #主体
        self.roberta = RobertaModel(config,add_pooling_layer=False)
        #头部
        self.dropout = nn.Dropout(config.hidden_dropout_prob)
        self.classifier = nn.Linear(config.hidden_size,config.num_labels)

        #初始化参数
        self.init_weights() #pretrainedmodel

    def forward(self,input_ids=None,attention_mask=None,token_type_ids=None,labels=None,**kwargs):
        outputs = self.roberta(input_ids,attention_mask,token_type_ids,**kwargs)#得到编码器的隐藏层表示
        sequence_output = self.dropout(outputs[0])#last_hidden_state
        logits = self.classifier(sequence_output)
        print(logits.shape)
        loss = None
        if labels is not None:
            loss_fct = nn.CrossEntropyLoss()
            loss = loss_fct(logits.view(-1,self.num_labels),labels.view(-1))
        return TokenClassifierOutput(loss,logits,hidden_states=outputs.hidden_states,attentions=outputs.attentions)
```

```python
""" 对整个数据集进行标签id与标记对齐 """
# tokenized_input = xlmr_tokenizer(panx_ch['de']['train']['tokens'],is_split_into_words=True)
# tokenized_input.word_ids()
def tokenize_and_align_labels(examples):#此处的examples指的是train/valid/test等dataset数据
    tokenized_input = xlmr_tokenizer(examples['tokens'],truncation=True,is_split_into_words=True)
    labels = []
    for idx,label in enumerate(examples['ner_tags']):#此处的label是数字标记,对于每一批样本
        word_ids = tokenized_input.word_ids(batch_index=idx)
        previous_word_idx=None
        label_ids=[] #用来存储样本中的每个token对应的lableid
        for word_idx in word_ids:
            if word_idx is None or word_idx==previous_word_idx:
                label_ids.append(-100)#特殊字符或和前一个字符相等，都没有label，用-100
            else:
                label_ids.append(label[word_idx])#label是当前样本的标记，对于类别相同的token,其标记是相同的(包括分割后的token)
            previous_word_idx=word_idx
        labels.append(label_ids)#添加每一个样本的标签
    tokenized_input['labels']=labels #增加新的列
    return tokenized_input
#对一整个数据集进行对齐，DatasetDict类型：包括了train,valid,test
def align_panx_dataset(dataset):
    return dataset.map(tokenize_and_align_labels,batched=True,remove_columns=['langs','ner_tags','tokens'])
```

```python
""" 要将输出的预测和实际标签由id转为对应的标签值才能使用seqeval的评分函数 """
import numpy as np
def align_prediction(prediction,label_ids): #prediction为输出的向量h,
    preds = np.argmax(prediction,axis=2)
    batch_size,seq_len = preds.shape
    label_list,pred_list = [],[]#存储每个样本标签值
    for batch_idx in range(batch_size):#处理每一个样本
        example_labels,example_pred = [],[]#存储样本中每个词的标签
        for seq in range(seq_len):
            if label_ids[batch_idx][seq]!=-100:
                example_labels.append(index2tag[label_ids[batch_idx][seq]])
                example_pred.append(index2tag[preds[batch_idx][seq]])
        label_list.append(example_labels)
        pred_list.append(example_pred)
    return pred_list,label_list
```

```python
"""
错误分析 
    查看损失最大的验证集例子
 """
from torch.nn.functional import cross_entropy
def forward_pass_with_lable(batch):
    #convert dict of lists to list of dicts suitable for data collator
    features = [dict(zip(batch,t))for t in zip(*batch.values())]
    batch = data_collator(features)
    input_ids = batch['input_ids'].to(device)
    attention_mask = batch['attention_mask'].to(device)
    labels = batch['labels'].to(device)
    with torch.no_grad():
        output = model(input_ids,attention_mask)
        predicted_label = torch.argmax(output.logits,axis=-1).cpu().numpy()
    loss = cross_entropy(output.logits.view(-1,7),labels.view(-1),reduction='none')
    loss = loss.view(len(input_ids),-1).cpu().numpy()
    return{"loss":loss,"predicted_label":predicted_label}
```

```python

def get_samples(df):
    for index,row in df.iterrows():#每一行是一个序列
        labels,preds,tokens,loss = [],[],[],[]
        for i,mask in enumerate(row['attention_mask']):
            if i not in{0,len(row['attention_mask'])}:#是两个数不是范围
                labels.append(row['labels'][i])
                preds.append(row['predicted_label'][i])
                tokens.append(row['input_tokens'][i])
                loss.append(f"{row['loss'][i]:.2f}")
        df_tmp = pd.DataFrame({'tokens':tokens,"labels":labels,'preds':preds,'loss':loss}).T#是对每个序列统计损失
        yield df_tmp
df['total_loss'] = df['loss'].apply(sum)
df_tmp = df.sort_values(by='total_loss',ascending=False).head(3)
for sample in get_samples(df_tmp):
    display(sample)    
```
